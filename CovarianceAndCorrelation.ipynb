```python
%matplotlib inline

import numpy as np
from pylab import *

def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

page_speeds = np.random.normal(3.0, 1.0, 1000)
purchase_amount = np.random.normal(50.0, 10.0, 1000)

scatter(page_speeds, purchase_amount)
covariance (page_speeds, purchase_amount)
```

```python
purchase_amount = np.random.normal(50, 10, 1000) / page_speeds

scatter(page_speeds, purchase_amount)

covariance(page_speeds, purchase_amount)
```

```python
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x, y) / stddevx / stddevy

correlation(page_speeds, purchase_amount)
```

```python
np.corrcoef(page_speeds, purchase_amount)
```

```python
purchase_amount = 100 - page_speeds * 3

scatter(page_speeds, purchase_amount)

correlation (page_speeds, purchase_amount)
```

```python
np.corrcoef(page_speeds, purchase_amount)
```
